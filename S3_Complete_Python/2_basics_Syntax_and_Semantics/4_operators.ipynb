###Arithmetic operators
Addition +
Subtraction -
Multiplication *
Division /
Floor Division // (Divisao inteira)
Modulus  % (Resto da divisao)
Exponentiation **

```python
restoDivisao=20//3
print(restoDivisao)

modulo10=10%4
print(modulo10)

#Exponentiation
exponentiation=2**10
print(exponentiation)
```

###Comparison operators
Equal to
Not equal to
Greater than
Less than
Greater than or equal to
Less than or equal to

```python
#Equal to (==)
a=10
b=20
print(a==b)

str1="Krish"
str2="Krish"
print(str1==str2)


#Not equal to (!=)
print(str1!=str2)


#Greather than (>)
print(a>b)


#Less than (<)
print(a<b)


#Greather than or equal to(>=)
print(a>=b)
print(a>=10)


#Less than or equal to(<=)
print(a<=b)
print(a<=5)
```

###Logical operators
AND
OR
NOT

```python
x=True
y=False
z=True

#AND
boolean1= x and y
print(boolean1)

#OR
boolean2= x or y or z
print(boolean2)

#NOT
boolean2= not x
print(boolean2)
```
