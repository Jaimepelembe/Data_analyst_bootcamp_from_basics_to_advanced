```python
#Understanding variable types
#Python is dynamically typed. The type of a variable is determined at runtime
age=25
```

```python
#Type Checking and conversion
type(age)
print(type(age))


#Type conversion
ageString=str(age) #Converts a object in to a string
print(type(ageString))

height=5.11
print(type(height))
height=int(height) #Converts the float value in a integer
print(type(height))
```

```python
#Dynamic Typing
#Python allows the type of a variable to change as the program executes.
var=10 #int
print(var, type(var))

var="Hello"
print(var, type(var))

var = 3.14
print(var, type(var))
```

```python
#Basic calculator
while True:
    inputString=input("Insert number one?")
    numberOne=float(inputString)
    inputString=input("Insert number two?")
    numberTwo=float(inputString)

    print("Choose the operation: \n 1. Addition \n 2. Subtraction \n 3. Multiplication \n 4. Division ")
    inputString=input()
    operation=int(inputString)

    match operation:
        case 1:
            operationResult= numberOne+numberTwo
            print(operationResult)

        case 2:
            operationResult=numberOne-numberTwo
            print(operationResult)
        case 3:
            operationResult=numberOne*numberTwo
            print(operationResult)
    
        case 4:
            if numberOne !=0:
                operationResult=numberOne/numberTwo
                print(operationResult)
            else:
                print("The numerator is 0. And it is impossible to divide by zero")
            
    print("Do you want to continue? \n 1.Yes \n 2. No")
```
