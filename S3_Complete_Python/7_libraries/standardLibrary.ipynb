**Data Serialization**
É o processo de converter dados ou objectos em memória para um formato padrão, de modo que possam ser armazenados, transmitidos ou reconstruídos mais tarde.

Exemplos de formatos de serialização:
JSON (muito usado, legível)
XML
YAML
CSV (estruturas simples)
Binary formats (Protocol Buffers, MessagePack)
Pickle (específico do Python)

```python
import json

#Convert a object into Json
data={"name":"Krish", "age":36}
jsonStr=json.dumps(data) # Converte o objecto Python numa string no formato JSON
print(jsonStr)
print(type(jsonStr))

#Convert from Json to a Python object
print("")
print("Deserialization: ")
dictionary=json.loads(jsonStr)
print(dictionary)
print(type(dictionary))
```

CSV

```python
import csv

#Write in a CSV file
with open("example.csv", mode="w", newline="") as file:
    writer=csv.writer(file)
    writer.writerow(["name","age"])
    writer.writerow(["Krish","32"])

#Read from a CSV file
with open("example.csv", mode="r") as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

Datetime

```python
from datetime import datetime,timedelta

now=datetime.now()
print(now)
```

```python
delta=timedelta(days=5)
print(delta)


#Somar tempo a uma data
in10days=now+timedelta(days=10)
print(in10days)


#Subtrair tempo a uma data
yestarday=now-timedelta(days=1)
yestarday

#Difference between two dates
date1=datetime(2026,1,1)
date2=datetime(2026,1,10)
differenceDates=date2-date1
print(differenceDates)
firstTime=datetime.now()
```

```python
#Elapsed time
elapsedTime=datetime.now()-firstTime
print(elapsedTime)
```
