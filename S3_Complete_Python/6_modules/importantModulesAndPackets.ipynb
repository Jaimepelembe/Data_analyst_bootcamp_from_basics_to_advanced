Import an internal package

Syntax: import packageName

```python
#Import the complete package
import math
math.sqrt(16)
```

```python
#Import parts of a package
#Syntax: from packageName import partsName

from math import sqrt,pi
print(sqrt(16))
print(sqrt(20))
print(pi)
```

```python
#Import with alias (nickname)
import math as mt
print(mt.sqrt(25))
```

```python
#Import all functions(*), variable and etc.
from math import *
print(cos(0))
```
