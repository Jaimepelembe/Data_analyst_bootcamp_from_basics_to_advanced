Sets

Introduction
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order.
Sets are useful for memebership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.



```python
# Creating a set
mySet={1,2,3,4,5}
print(mySet)
print(type(mySet))

#Empty set
emptySet={}
print(emptySet)

#Convert a list into a set
listNumbers=[1,2,3,4,1,2,3,4]
setNumbers=set(listNumbers)
print(setNumbers)
```

Basics sets operation

```python
# Adding and removing 

#Adding
print(mySet)
mySet.add(6)
mySet.add(7)
print(mySet)

#Removing
mySet.remove(6)
mySet.remove(4)
print(mySet)

#Discarding
mySet.discard(6)
print(mySet)



```

```python
#Pop method
removedElement=mySet.pop()
print(removedElement)
print(mySet)
```

```python
#Clear()
#Clear all the elements
mySet.clear()
print(mySet)

```

Set membership test
Verify if an element is a member of the set or not.

```python
setAges={18,19,20,22,24}
print(19 in setAges) 
print(25 in setAges)
```

Mathematical operation

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

#Union
unionSet=set1.union(set2)
print("Union set: ",unionSet)


#Intersection
intersectionSet=set1.intersection(set2)
print("Intersection set: ",intersectionSet)


#Intersection with update
set1.intersection_update(set2) #Update the value of the Set1 with the intersection between set1 and set2
print("Set 1 updated: ",set1)
```

```python
#Difference
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
differenceSet=set1.difference(set2)
print(differenceSet)

#Difference with update
#set1.difference_update(set2)
```

```python
#Symmetric difference
#Removes the symmetric elements
symmetricDifference=set1.symmetric_difference(set2)
print(symmetricDifference)
```

Set Methods

```python
set1={1,2,3,4,5}
set2={3,4,5}


# Is subset
print(set1.issubset(set2))


# Is superset
print(set1.issuperset(set2))
```

```python
#Examples

#Removes the duplicates from a list
duplicatedList=[1,2,2,3,4,4,5]
simpleList=set(duplicatedList)
print(simpleList)
```

```python
# Counting unique words in text
text="In this tutorial we are discussing about sets. We hope that you will have a good experience"
words=text.split()
print(words)

#Convert list of words to set to get unique words
uniqueWords=set(words)
print("Unique words: ", uniqueWords)
print(len(uniqueWords))
```
