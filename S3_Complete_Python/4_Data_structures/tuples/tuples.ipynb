Tuples in Python

Introduction to tuples
Tuples are ordered collection of items that ae immutable. They are similar to lists, but their immutability makes them different.

```python
#Creating a tuple
emptyTuple=()
print(emptyTuple)
print(type(emptyTuple))

#Second method
tuple2=tuple()
print(type(tuple2))
```

```python
#Adding elements 
numbers=tuple([1,2,3,4,5,6]) # A list can be converted into a tuple and vice versa
print(numbers)
listNumbers=list(numbers)
print(listNumbers)
```

```python
#Mixed tuple
# The tuples allows different data types

mixedTuple=(1,"Hello Word", 3.14, True)
print(mixedTuple)
print(mixedTuple[1:])
```

```python
#Acessing tuples elements
#Using the index
print(numbers[1])
print(numbers[-1]) #The last element
print(numbers[::-1]) #Gets all elements in the reverse order
```

Tuples operations

```python
#Concatenation of tuples (+)
concatenatedTuple=numbers+mixedTuple
print(concatenatedTuple)
```

```python
#Multiplication of elements (*)
multiplicatedNumbers=numbers*3
print(multiplicatedNumbers)
```

```python
# Immutable nature of tuples
# Tuples are immutable, meaning their elements can not be changed once assigned.

list1=[1,2,3,4,5]
print(list1)
list1[0]=26

#Tuples
numbers[0]="Rango"
# You can convert the tuple in a list, make the assignment and then convert the list in a tuple.
```

Tuple methods

```python
#Count
#Count the number of occurance of an element in the tuple
numbers2=(1,2,1,2,3,2,1,3)
print(numbers2.count(2))

#index
#Returns the index of the first occurance of an element in the tuple
print(numbers2.index(2))


```

Packing and unpacking tuple

```python
#Packing
packedTuple=1,"Hello", 3.14 
print(packedTuple)

#Unpacking
a,b,c=packedTuple
print(a)
print(b)
print(c)

#Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
 
*first2,middle2=numbers
print(first2)
print(middle2)
```
