Example 1: Manage a to do list

Create a to do list to keep tack of tasks

```python
#In The to do list you need to be able to add, update and delete a task
todoList=["Buy groceries", "Pay bills","Clean the house"]

#Adding a task
todoList.append("Schedule meeting")
todoList.append("Go For a Run")


#Removing a completed task
todoList.remove("Clean the house")

#Checking if a task is in the list
if "Pay bills" in todoList:
    print("Don't forget to pay the bills")

print("To do list remaning")
for task in todoList:
    print(f"-{task}")
```

Example 2: Organizing student grades
Create a list to store and calculate average grades for students

```python
# Organizing student grades
grades = [85,92,78,90,88]

#adding a new grade
grades.append(95)

#Calculating the average grade
averageGrade =sum(grades)/ len(grades)
print(f"Average Grade: {averageGrade:.2f}")

#Finding the higest and lowest grades
highestGrade=max(grades)
lowestGrade=min(grades)
print(f"Higest Grade: {highestGrade}")
print(f"lowest Grade: {lowestGrade}")
```

Example 4: Collecting user Feedback

Use a list to collect and analyze user feedback.

```python
#Collecting user feedback
feedback =["Great service!", "Very satisfied", "Coud be better", "Excellent experience"]

#Adding new feedback
feedback.append("Not happy with the service")

# Counting specific feedback
positiveFeedbackCount= sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())
print(f"Positive feedback count: {positiveFeedbackCount} \n")

#Print all feedback
print("Feedbacks:")
for comment in feedback:
    print(comment)
```
