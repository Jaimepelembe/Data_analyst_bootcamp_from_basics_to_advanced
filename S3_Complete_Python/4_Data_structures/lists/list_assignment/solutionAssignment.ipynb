Assignment 4: List Comprehensions

Create a new list containing the squares of the first 10 positive integers using a list comprehension. Print the new list.


```python
listNumbers=[i**2 for i in range(10)]
print(listNumbers)
```

Assignment 5: Filtering Lists

Create a new list containing only the even numbers from the list created in Assignment 1 using a list comprehension. Print the new list.

```python
listEvenNumbers=[i for i in listNumbers if i%2 !=0]
print(listEvenNumbers)
```

Assignment 6: List Methods

Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

```python
import random 
listRandomNumbers=[random.randint(0,20) for i in range(20)]
```

```python
print(f"Original list: {listRandomNumbers}")
listRandomNumbers.sort()
print(f"List sorted in ascending order: {listRandomNumbers}")
listRandomNumbers.sort(reverse=True)
print(f"List sorted in descending order: {listRandomNumbers}")
```

```python
#Removing duplicates
for item in listRandomNumbers:
    print(item)
    counter=listRandomNumbers.count(item)
    while counter>1:
        listRandomNumbers.remove(item)
        counter=counter-1

print(f"The final list: {listRandomNumbers}")
        
```
