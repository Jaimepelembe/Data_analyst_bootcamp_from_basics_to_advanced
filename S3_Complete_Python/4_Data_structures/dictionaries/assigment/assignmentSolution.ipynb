Assignment 10: Dictionary and List Conversion
Create a dictionary with the first 5 positive integers as keys and their squares as values. Convert the dictionary to a list of tuples and print it.

```python
dictionarySquares={x:x**2 for x in range(5)}
print(dictionarySquares)

listTuple=[]
for key,value in dictionarySquares.items():
    listTuple.append((key,value))
print(listTuple)
#listTuple2=[(key,value) for key, value in dictionarySquares.items()]  List comprehension
#print(listTuple2)
```

Assignment 13: Default Dictionary
Create a default dictionary where each key has a default value of an empty list. Add some elements to the lists and print the dictionary.

```python
#In Python, defaultdict is a subclass of the built-in dict class from the collections module. 
#It automatically assigns a default value to keys that do not exist, which means you donâ€™t have to manually check for missing keys and avoid KeyError.

from collections import defaultdict
defaultDictionary=defaultdict(list)

defaultDictionary["fruits"].append("Apple")
defaultDictionary["fruits"].append("Orange")
defaultDictionary["vegetables"].append("Carrot")
defaultDictionary["vegetables"].append("corn")
print(defaultDictionary)
print(defaultDictionary["fruits"])
print(defaultDictionary["meat"])
```

Assignment 15: Dictionary and JSON

Create a dictionary representing a book with keys 'title', 'author', 'year', and 'genre'. Convert the dictionary to a JSON string and print it.

```python
import json

dictionaryBook={
"title":"Atomic Habits",
"author":"James Clear",
"year":"2019",
"genre":"Business Management"
}

print("Dictionary: ", dictionaryBook)

#Convert the dictionary into JSON
jsonElement=json.dumps(dictionaryBook)
print("Json: ", jsonElement)
```
