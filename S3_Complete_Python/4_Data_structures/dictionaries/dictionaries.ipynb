Introduction

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
#Creating
emptyDictionary={}
print(type(emptyDictionary))

emptydictionary2= dict()
print(type(emptyDictionary))
```

```python
dictStudent={"name":"Krish", "age":32, "grade":24}
print(dictStudent)
print(type(dictStudent))
print(dictStudent["age"])
```

```python
#Single key is always used
#Repeating the key
student={"name": "Thanos", "age":3, "name":"Zangau"}
print(student)
```

Accessing dictionary elements

```python
#Using the key
print(student["age"])


#Using get(method)
print(student.get("name"))
print(student.get("grade")) #There is no this key, so the dictionary will return none.
print(student.get("grade","Not avaliable")) # Defining the default value to return in case of a lack of key
```

Modifying elements
Dictionary are mutable.

```python
print(student)

#Updating the value for the key
student["age"]=18
print(student)


#Adding a new key and value
student["grade"]=19
student["curse"]="Data analysis"
print(student)
```

```python
#Deleting key and value pair
del student["curse"]
print(student)
```

Dictionary methods

```python
#Get all the keys or values
keys=student.keys()
print(keys)
values=student.values()
print(values)


#Get the pair key-value
pairs=student.items()
print(pairs)
```

```python
#Copying a dictionary
studentCopy=student # Wrong way to copy, because python will point to the same memory location
print("Original: ", student)
print("Copy: ", studentCopy)
```

```python
student["name"]="Kira"
print(student)
print(studentCopy)
```

```python
studentCopyCorrect=student.copy() #Shallow copy provides a different memory location for the new variable(dictionary)

student["name"]="Ana Maria"
print(student)
print(studentCopyCorrect)
```

Itarating over dictionaries

```python
#You can use loops to iterate over dictionaries, keys, values or items.

#Iterating over keys
for key in student.keys():
    print(key)    
```

```python
#Iterating over values
for value in student.values():
    print(value)
```

```python
#Iterating over items
for key, value in student.items():
    print(key,":",value)
```

Nested dictionaries

In No SQL DB such as Mongo DB everthing is stored in key-value pairs

```python
students={
"student1":{"name":"Ana", "age":25},
"student2":{"name":"Maria", "age":20},
"student3":{"name":"Fatima", "age":18},
"student4":{"name":"Cisco", "age":24}
}

print(students)

#Acessing nested dictionaries
print(students["student1"]["name"])
print(students["student1"]["age"])
```

```python
#Iterating over nested dictionaries
for studentID, studentInfo in students.items():
    print(f"{studentID}:{studentInfo}") 
    for key, value in studentInfo.items():
        print(f"{key}:{value}")
    print("")
        
```

```python
#Dictionary comprehension
squares={ x:x**2 for x in range(5)} #Dictionary of squares
print(squares)
```

```python
squaresEvenNumbers={x:x**2 for x in range(10) if x%2 !=0} #Squares of even numbers
print(squaresEvenNumbers)
```

```python
#Practical examples
#1. Use a dictionary to count the frequency of elements in list
dictionaryFrequency={}
numbers=[1,2,2,3,3,3,4,4,4,4,5,6,7,5,7,7,5,2,1,6]
for i in numbers:
    if i not in dictionaryFrequency.keys():
        dictionaryFrequency[i]=numbers.count(i)

print(dictionaryFrequency)
```

```python
#Option two of solution
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
#Merge two dictionaries into one

#Using duble asterisk
dictionary1={"a":1,"b":2}
dictionary2={"b":3,"c":4}
mergedDictionary={**dictionary1,**dictionary2} # **dictionary1 -> desempacota todas as chaves e valores de dictionary1 e faz o mesmo com o dictionary2. Se houver chaves repetidas, o valor do ultimo dicionario prevalece
print(mergedDictionary)
```

```python
#Using a bar (|) for python 3.9+
mergedDictionary2= dictionary1 | dictionary2
print(mergedDictionary2)
```

```python
#Actualizar um dicionario existente
print(dictionary1)
dictionary1.update(dictionary2)
print(dictionary1)
```
