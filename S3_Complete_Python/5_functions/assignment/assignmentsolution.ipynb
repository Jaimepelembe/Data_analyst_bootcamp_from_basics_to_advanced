Assignment 6: Nested Functions
Define a function that contains another function inside it. The outer function should take two integers as input and return the result of the inner function, which multiplies the two integers. Test the function with different inputs.

```python
def outerFunction(a,b):
        
    def innerFunction(x,y):
        return x*y

    return innerFunction(a,b)

print(outerFunction(2,4))
print(outerFunction(5,4))
```

Assignment 8: Recursive Function
Define a recursive function that calculates the fibonacci of a given number. Test the function with different inputs.
Base case:
FB(0)=1
FB(1)=1

Recursive cse:
FB(2)=F(1)+F(0)=2
FB(3)=F(2)+F(1)=3
FB(4)=F(3)+F(2)=5

```python
def fibonacci(number):
    """Calculate the fibonacci of a given number"""
    if number==0 or number==1: # Base case
        return 1
    else: # Recursive case
        return fibonacci(number-1)+fibonacci(number-2)


print(fibonacci(2))
print(fibonacci(4))
print(fibonacci(5))
    
```
