Introduction

Map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator).

This is particularly useful for transforming data in a list comprehensively.

```python
def square(number):
    return number**2


square(4)
```

```python
# Calculate the square of all number in a list
numbers=[1,2,3,4,5,6]
squareNumbers=list(map(square,numbers))# We can use lambda function: map(lambda x:x**2, numbers)
```

```python
#Lambda function with map
evenNumbers=list(map(lambda number:number%2==0, numbers))
print(evenNumbers)
```

Map multiple iterables

```python
numbers1=[1,2,3]
numbers2=[4,5,6]
addedNumbers=list(map(lambda x,y: x+y, numbers1, numbers2))
print(addedNumbers)
```

Convert list of strings to integers
Use map to convert strings to integers


```python
strNumbers=['1','2','3','4','5','6']
print(strNumbers)
intNumbers=list(map(int,strNumbers))
print(intNumbers)
```

```python
words=["Apple", "Banana","Cherry"]
upperWords=list(map(str.upper,words))
print(upperWords)
```
