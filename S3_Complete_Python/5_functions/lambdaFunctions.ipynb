Introduction

Lambda functions are small anonymous functions definded using the lambda keyword. They can have any number of arguments but only one expression.
They are commonly used for short operations or as arguments to higher-order functions.

```python
# Syntax
lambda arguments : expression
```

```python
addition= lambda a,b: a+b

type(addition)
print(addition(5,6))
```

```python
even=lambda number: number%2==0
print(even(10))
print(even(15))
```

```python
#Lambda functions with multiple parameters

def additionThree(x,y,z):
    return x+y+z

additionThree(2,5,9)


lamdaAdditionThree= lambda x,y,z: x+y+z
print(lamdaAdditionThree(2,5,11))
```

Map function

Applies a function to all items in a collection of items, like a list, dictionary.

Syntax: map(function, iterable)

```python
numbers=[1,2,3,4,5,6]
squares=list(map(lambda x:x**2, numbers)) # We need to convert into a list, because map method returns a map object.
print(squares)
```
