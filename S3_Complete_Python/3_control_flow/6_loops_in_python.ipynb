```python
# 1. For loop
for i in range(5):
    print(i)
```

```python
for i in range(1,6) :
    print(i)


#Decrescent order
print("===========================")
for i in range(10,1,-1):
    print(i)
```

```python
#Interating Strings
string="Manchester City"

for i in string:
    print(i)
```

```python
#2.While loop

count=0

while count<5:
    print(count)
    count=count+1
```

```python
#3. Loop control statements

#3.1 Break
# The loop ends immediately and the program jump to the next instruction before the loop

for i in range (1,6):
    if i%4==0 and i<5:
        break
    print(i)
```

```python
#3.2 Continue
# Skips the current iteration and continues with the next iteration.

for i in range(15):
    if i%2==0:
        continue #Skips the odd numbers
    print(i)
```

```python
#3.3 Pass
# Is a null operation. It does nothing
#It works as a placeholder to keep a valid structure of code

if True:
    pass # without pass, Python would give an error, because it needs a block of code.

for i in range(5):
    pass 


for i in range(7):
    if i==3:
        pass
    print(i)
```

```python
#4. Nested loops
# A loop inside anoter loop

for i in range(3):
    for j in range(2):
        for k in range (2): 
            print (f"i: {i} and j: {j} and k: {k}")
```

```python
#Examples

#1. Calculate the sum of first N natural numbers using a while and for loop

inputString=input("How many first natural numbers do you want to sum?: ")
if inputString.isnumeric:
    naturalNumber=int(inputString)
    count=0
    sum=0
    while count<=naturalNumber:
        sum=sum+count
        count=count+1

    print(f"The sum of the firts {naturalNumber} natural number is: {sum}")
```

```python
#Using for loop:
inputString=input("How many first natural numbers do you want to sum?: ")
if inputString.isnumeric:
    naturalNumber=int(inputString)
    sum=0
    for i in range(naturalNumber+1):
        sum=sum+i
    
    print(f"The sum of the firts {naturalNumber} natural number is: {sum}")
```

```python
#Using for loop
limiteSuperior=1000
isPrime=True
print(f"The prime numbers between 1 and {limiteSuperior} are: ")
for i in range(2,limiteSuperior+1):
    for j in range (2,i): 
        if i%j==0:  # A prime number is only divided by one and it self. If it find another number it is not a prime number 
            isPrime=False
            break
        
    if isPrime: 
        print(i)
    else:
        isPrime=True
```
