```python
import pandas as pd
import numpy as np
```

```python
#1. Create a Pandas DataFrame with 4 columns and 6 rows filled with random integers. Set the index to be the first column.
import random

def dictionaryList(listKeys,rows):
    """Create a dictionary of a list
        Args:
            listKeys: List of the dictionary keys
            rows: Number of rows

        Returns:
        A dictionary
    """

    try:
        dictionary={}

        for key in listKeys:
            dictionary[key]=[]
            for row in range(rows):
                dictionary[key].append(random.randint(0,20))

    except TypeError as ex:
        print(ex)

    except KeyError as ex:
        print(ex)


    else:
        return dictionary



index=['a','b', "c","d"]
rows=6
data=dictionaryList(index,rows)
data

dataFrame=pd.DataFrame(data)
dataFrame.head()
```

```python
#Set the index to be the first column.

dataFrame.set_index("a", inplace=True)
dataFrame.head()
```

```python
#1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.
import math
dataFrame5=pd.DataFrame(np.random.randint(0,20, size=(5,3)))
print(dataFrame5,"\n")

#Introducting NaN values
print(dataFrame5[0][0])

dataFrame5.loc[0,0]=math.nan
dataFrame5.loc[1,2]=math.nan
dataFrame5.loc[2,2]=math.nan
print(dataFrame5[0][0])

print(dataFrame5)
```

```python
#Cleaning the dataFrame
#Verify how many NaN we have
dataFrame5.isnull().sum()
```

```python
#Fill the NaN values with the mean of the respective columns.
dataFrame5=dataFrame5.fillna(dataFrame5.mean(axis=0))
dataFrame5
```


**Assignment 4: Data Aggregation**

```python
#1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers.
#Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.
dataFrame6=pd.DataFrame(np.random.choice(["A","B","C"],size=10),columns=["Category"])
dataFrame6["Value"]=pd.DataFrame(np.random.randint(10,50, size=(10,1)))
dataFrame6
```

```python
#Compute the sum and mean of each column
groupedData=dataFrame6.groupby("Category")["Value"].agg(["sum","mean"])
groupedData
```

```python
#2. Create a Pandas DataFrame with 3 columns: 'Product', 'Category', and 'Sales'. Fill the DataFrame with random data.
#Group the DataFrame by 'Category' and compute the total sales for each category.
dataFrame7=pd.DataFrame(np.random.choice(["Kiwi","Casava","Potato","Watermelon","Sugar","Orange"],size=(18,1)) ,columns=["Product"])
dataFrame7["Category"]=pd.DataFrame(np.random.choice(["A","B","C"],size=(18,1)))
dataFrame7["Sales"]=pd.DataFrame(np.random.randint(2,10, size=(18,1)))
dataFrame7
```

```python
#Group the DataFrame by 'Category' and compute the total sales for each category.
groupedDataCategory=dataFrame7.groupby("Category")["Sales"].sum()
groupedDataCategory
```

**pandas.concat()** function concatenate two or more pandas objects like DataFrames or Series along a particular axis. It is especially useful when combining datasets either vertically (row-wise) or horizontally (column-wise).

pd.concat  cola DataFrames por linhas ou colunas, sem olhar para valores-chave

pd.merge  junta DataFrames com base numa chave (como um JOIN em SQL)


```python
# 2. Create two Pandas DataFrames with different columns. Concatenate the DataFrames along the rows and along the columns.
dataFrame1=pd.DataFrame(np.random.choice(["Meat","Bread","Fish"],size=(6,1)), columns=["Product"])
dataFrame1["Price"]=pd.DataFrame(np.random.randint(1,100,size=(6,1)))
dataFrame1

dataFrame2=pd.DataFrame(np.random.choice(["Orange","Apple","Avacato"], size=(6,1)), columns=["Fruits"] )
dataFrame2["Sales"]=pd.DataFrame(np.random.randint(1,10,size=(6,1)))
dataFrame2

print(dataFrame1,"\n")
print(dataFrame2)

```

```python
 #Concatenate the DataFrames along the rows and along the columns.
#Row-wise
concatenatedRows=pd.concat([dataFrame1,dataFrame2],axis=1)
concatenatedRows
```

```python
#Column-wise
concatenatedColumns=pd.concat([dataFrame1,dataFrame2])
concatenatedColumns
```
