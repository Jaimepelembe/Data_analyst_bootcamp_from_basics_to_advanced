```python
import pandas as pd
import numpy as np
```

```python
#1. Create a Pandas DataFrame with 4 columns and 6 rows filled with random integers. Set the index to be the first column.
import random

def dictionaryList(listKeys,rows):
    """Create a dictionary of a list
        Args:
            listKeys: List of the dictionary keys
            rows: Number of rows

        Returns:
        A dictionary
    """

    try:
        dictionary={}

        for key in listKeys:
            dictionary[key]=[]
            for row in range(rows):
                dictionary[key].append(random.randint(0,20))

    except TypeError as ex:
        print(ex)

    except KeyError as ex:
        print(ex)


    else:
        return dictionary



index=['a','b', "c","d"]
rows=6
data=dictionaryList(index,rows)
data

dataFrame=pd.DataFrame(data)
dataFrame.head()
```

```python
#Set the index to be the first column.

dataFrame.set_index("a", inplace=True)
dataFrame.head()
```

```python
#1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.
import math
dataFrame5=pd.DataFrame(np.random.randint(0,20, size=(5,3)))
print(dataFrame5,"\n")

#Introducting NaN values
print(dataFrame5[0][0])

dataFrame5.loc[0,0]=math.nan
dataFrame5.loc[1,2]=math.nan
dataFrame5.loc[2,2]=math.nan
print(dataFrame5[0][0])

print(dataFrame5)
```
