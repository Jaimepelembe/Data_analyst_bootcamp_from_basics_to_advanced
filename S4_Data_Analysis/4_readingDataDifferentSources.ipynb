**Reading Data from Different Sources**

```python
import pandas as pd
```

**Json file**

```python
from io import StringIO

data='{ "employee_name": "James" , "email":"james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}] }'
dataFrame=pd.read_json(StringIO(data))
dataFrame
```

```python
dataFrame["email"]
```

```python
#Convert from DataFrame to Json

jsonFile=dataFrame.to_json()# The parameter orient by default is the index of the dataFrame
jsonFile
```

```python
#Defining an orient value
#Orient=index
jsonFile2=dataFrame.to_json(orient="index")
jsonFile2

```

```python
#Orient=records
jsonFile3=dataFrame.to_json(orient="records")
jsonFile3
```

**Using url**

```python
dataFrame2=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)
dataFrame2.head()
```
