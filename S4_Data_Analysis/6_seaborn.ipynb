# Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

# Basic Plotting

## 1. Loading the dataset

```python

tips=sns.load_dataset('tips')
tips.head()
```

## 2. Pre-processing the dataset

```python
# Check the data types of each column
print("Types of each column:")
print(f"{tips.dtypes}\n")


print("Number of null values:")
print(f"{tips.isnull().sum()}\n")
```

```python
# Check statistical information
tips.describe()
```

## 3. Plotting the graphics

### 3.1 Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bills vs Tip", pad=25)
plt.show()
```

### 3.2 Line plot

```python
sns.lineplot(x='size', y='total_bill', data=tips, marker='o')
plt.title(" Size vs Total bill")
plt.grid()
plt.show()
```

### 3.3 Categorical plots

```python
# Bar plot
sns.barplot(x="day", y="total_bill",data=tips)
plt.title("Day vs Total bill", pad=20)
plt.grid()
plt.show()
```

```python
#Box plot

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Day vs Total bill") 
plt.show()
```

```python
# Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Day vs Total bill", pad=25) 
plt.show()
```

### 3.4 Histograms

```python
sns.histplot(tips["total_bill"], bins=10, kde=True) #kde= True -> Shows the really shape of the histogram using a curve
plt.title("Total bills")
plt.show()
```

### 3.5 KDE Plot

```python
sns.kdeplot(tips["total_bill"], fill=True)
```

### 3.6 Pair plot

```python
sns.pairplot(tips) # Shows the relation between all the variables to different ways. Only uses numerical variables
```

# 3.7 Heatmap

```python
correlation=tips[["total_bill", "tip","size"]].corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap="coolwarm")
```

### Make some customizations 

```python
import pandas as pd

salesDf= pd.read_csv("salesData.csv")
salesDf.head()
```

```python
# Plot total sales by product

plt.figure(figsize=(10,6))
sns.barplot(x="Category", y="Sales", data=salesDf, estimator=sum)
plt.title("Total sales by product", pad=25)
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Plot total sales by Region

plt.figure(figsize=(10,6))
sns.barplot(x="Region", y="Sales", data=salesDf, estimator=sum)
plt.title("Total sales by Region", pad=25)
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```
