```python
import pandas as pd
```

```python
df=pd.DataFrame({'nome': ['Ana', 'Carlos', 'João','Flora','Pedro','Tomas'],
    'cidade': ['Maputo', 'Niassa', 'Tete', 'Gaza', 'Inhambane','Tete']})

df1=df
df2=df
```

```python
# Using Map method
df1["cidade"]=df1["cidade"].map({"Maputo":0,"Niassa":1,"Tete":2,"Gaza":3, "Inhambane":4})

df1
```

```python
#Using LabelEncoder (Automatic)

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df2["cidade"]=le.fit_transform(df2["cidade"])
df2
```
