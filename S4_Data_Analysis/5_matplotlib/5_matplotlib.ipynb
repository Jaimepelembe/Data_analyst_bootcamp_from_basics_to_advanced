**Data Visualization With Matplotlib**

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

**Create a linear plot**

```python
x=[1,2,3,4,5]
y=[i**2 for i in x]

plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic line plot", loc="center", pad=20)
```

```python
#Create a customized line plot

plt.plot(x,y, color="red", linestyle="--", marker="o", linewidth=3, markersize=9) # Linestyle=[-, --, :,  -., 'None', ' ', '', 'solid'] 
plt.grid(True)
```

**Multiple plots**

**Scatter plot**

Um scatter plot  um grfico usado para mostrar a relao entre duas variveis numricas.
Cada ponto no grfico representa um par de valores (x, y).

Syntax: matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, alpha=None, edgecolors=None, label=None)

s = Marker size (scalar or array-like)
c = Marker color
cmap = Colormap for mapping numeric values to colors
alpha = Transparency (0 = transparent, 1 = opaque)
label = Legend label for the dataset




# Para que serve o scatter plot

Ele  muito usado para:

Detectar relaes entre variveis (positiva, negativa ou nenhuma);

Identificar padres;

Encontrar outliers (valores fora do normal);

Analisar dados experimentais e estatsticos;

Exemplo prtico:

Altura vs peso;

Horas de estudo vs nota;

Temperatura vs consumo de energia;



# Use quando:

Tens duas variveis numricas;

```python
# Example 3: Demonstrates how to customize a scatter plot using different marker sizes and colors for each point. Transparency and edge colors are also adjusted.

x=np.array([3,12,9,20,5,18,22,11,27,16])
y=np.array([95,55,63,77,89,50,41,70,58,83])

markerSize=[20,50,100,200,500,1000,60,90,150,300]
markerColor=['red','green','blue','purple','orange','black','pink','brown','yellow','cyan']

plt.scatter(x,y,s=markerSize,c=markerColor, alpha=0.6, edgecolors='w', linewidths=1)
plt.title("Scatter plot with varying colors and sizes", pad=25)
plt.grid()
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

**Pie chart**

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data.


Syntax: matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False) 


# Para que serve o pie chart

Comparar partes de um todo;

Mostrar percentagens;

Ter uma viso rpida da distribuio.

# Exemplos tpicos:

Distribuio de despesas mensais;

Percentagem de alunos por curso;

Tipos de produo (milho, arroz, feijo, etc.).


# Usa pie chart quando:

Tens poucas categorias (2 a 5);

O foco  proporo, no preciso.

#  Evita pie chart quando:

Tens muitas categorias;

As diferenas so pequenas;

Queres comparar valores entre grficos.

```python
# Example 1: 
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=["gold","yellowgreen","lightcoral","lightskyblue"]
explode=(0.2,0,0,0) # Move out the first slice

#Create a pie chart
percentageFormat="%1.1f%%"
plt.pie(sizes, labels=labels,colors=colors, autopct=percentageFormat, shadow=True, explode=explode) 
plt.show()
```

```python
# Example 2: Cars distribution

cars=["AUDI","BMW","FORD","TESLA","JAGUAR","MERCEDES"]

data=[23,17,35,29,12,41]
percentageFormat="%1.1f%%"
plt.pie(data,labels=cars, autopct=percentageFormat)
#plt.legend(title="Cars", loc="upper left")
plt.show()
```

# **Boxplot**

## O que  Boxplot

 uma ferramenta grfica que permite verificar a distribuio dos dados e os valores discrepantes (outliers).


### Percentil
Divide os dados em 100 partes iguais. 

"Que valor separa X% dos dados abaixo dele?"

Ex: 

Percentil 25 (P25)  25% dos dados esto abaixo desse valor

Percentil 50 (P50)  50% dos dados esto abaixo ( a mediana)

Percentil 90 (P90)  90% dos dados esto abaixo

### Quartil

Divide os dados em 4 partes iguais.

| Quartil |	Percentil |	Significado |
| :-------:|----------:|------------:|
|Q1	|P25	|25% dos dados abaixo|
|Q2	|P50	|Mediana|
|Q3|P75	|75% dos dados abaixo|

### Interpretao do Boxplot

O Boxplot nos fornece informao da posio, disperso, simetria, caudas e valores discrepantes.

#### Posio

Verifica-se a mediana Q2

#### Disperso

Intervalo interquartlico 3 quartil - 1 quartil ou pela amplitude = valor maximo - valor minimo

#### Simetria

Dados simetricos quando a linha da mediana est no centro do rectngulo (2 quartil);

Assimtricos positivos = linha mediana prxima a 1 quartil

Assimtricos negativos = linha mediana prxima a 3 quartil


#### Ouliers
Valores que esto a cima ou a baixo do limite de deteco


Examples

```python
# 1. Boxplot of ages
ages=[30, 31,34,34,37,37,42,42,45,45,45,46,46,48,50,52,54,54,54,80] 
plt.boxplot(ages)
plt.show()
print("Dados tem uma assimetria negativa")
```

```python
# Auxiliar functions

def removeNullValues(dataset,column):
    """Removes the null values from a column of the dataset"""

    try:
        dataset=dataset.dropna(subset=[column])
    
    except KeyError as key:
        print(f"The column {column} doesn't exist in the dataset.")
    
    except NameError as ne:
        print(f"The dataset {dataset} doesn't exist.")

    except Exception as ex:
        print(ex)
    
    else:
        return dataset


def printRemotionNullValues(dataset,column):
    print(dataset[column].isnull().sum())
    dataset=removeNullValues(dataset,column)
    print(dataset[column].isnull().sum())
    return dataset



def removeOutliers(dataset,column):
    """Removes the outliers from a column of the dataset.
    
    Returns the dataset without outliers in the column
    """
    try:
        q1,q3 = np.percentile(dataset[column], [25,75])
        print(f" Q1 and Q3 = {q1} : {q3}" )
        
        iqr= q3 - q1
        lower = q1-1.5*iqr
        upper = q3+1.5*iqr
        print(f"Lower and upper = {lower} : {upper}")

        cleanDataset=dataset[ (dataset[column] >= lower) & (dataset[column]<=upper)]

    except KeyError as key:
        print(key)
    
    except NameError as ne:
        print(ne)
    except Exception as ex:
        print(ex)

    else:
        return cleanDataset


#removeOutliers(athletesDataset,"Weight")

#athletesDataset
    
```

## 4. Remove Outliers using the Interquartile Range (IQR) Method

Remove Extreme values beyonde a reasonable range to improve model robustness.

IQR = Q3 (75th percentile) - Q1 (25th percentile)

lower (Min) = Q1 - 1.5 * IQR

upper (Max) = Q3 + 1.5 * IQR

Values below lower or above upper are outliers;
Calculate lower and upper bounds for each column separately;
Filter data points to keep onlythose withn bounds
